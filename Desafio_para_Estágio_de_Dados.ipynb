```
import pandas as pd
import numpy as np

#importar tabela
tabela= pd.read_csv("last-30d-facebook.csv")

#apagar linhas com valores vazios
tabela=tabela.dropna()

#apagar linha 5 que nao estava ajudando na visualizaçao
tabela=tabela.drop(5, axis=0)

#renomear titulos
tabela=tabela.rename(columns={'Unnamed: 0': 'Campanha','Unnamed: 2':'Evento Otimizador','Unnamed: 1':'Base Usuarios','Unnamed: 3':'Midia','Unnamed: 4':'Gasto','Unnamed: 5':'Ivestimento alocado','Unnamed: 6':'Clicks','Unnamed: 7':'Impressoes', 'Unnamed: 8':'Sessões','Unnamed: 9':'Conversao','Unnamed: 10':'Receita','Unnamed: 11':'Primeira compra'})

#tabela = tabela.loc[6:17].eval('Lucro = Receita - Gasto)

#print('exibir somente coluna de campanha, gasto, investimento alocado e receita')
display(tabela.iloc[:,[0,4,5,10]])

#exibir somente coluna de campanha, clicks,impressoes e conversao
display(tabela.iloc[:,[0,6,7,9]])

#Colocar a tabela em ordem crescente baseada na coluna clicks
tabela.sort_values("Clicks", axis = 0, ascending = True,inplace = True, na_position ='last')

#C	R$ 9.000	R$ 8.000	R$ 18.810
#E	R$ 1.000	R$ 1.500	R$ 9.476
#K	R$ 1.050	R$ 2.500	R$ 14.000
```
